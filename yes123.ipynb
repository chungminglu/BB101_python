```python
from splinter import Browser #sign in function
import splinter
browser = Browser('chrome')
url = "http://www.yes123.com.tw/admin/index.asp"
browser.visit(url)
browser.find_by_xpath('//*[@id="search_work"]').first.click()
browser.find_by_xpath('//*[@id="a2_1011_0000_0000"]').first.click()
browser.find_by_xpath('//*[@id="s2_1011_0001_0000"]').first.click()
browser.find_by_xpath('//*[@id="_lv2_menuList"]/tbody/tr[8]/td/label/ul').first.click()
browser.find_by_xpath('//*[@id="_lev1_bn_close"]').first.click()
browser.find_by_xpath('//*[@id="_idx_sh_01"]/span[4]/a').first.click()
```

```python
HOST = 'http://www.yes123.com.tw'
jobs_list = []
from bs4 import BeautifulSoup
for i in range(1,100):
    soup = BeautifulSoup(browser.html, 'lxml')
    jobs_section = soup.select('body > div.SLV_Wrapper > div.box_detail > div.box_detail_left')[0]
    jobs = jobs_section.select('div > ul > li > a')
    for job in jobs:
        jobs_list.append([job.text, HOST + job['href']])
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    browser.find_by_text('>').click()

```

```python
jobs_list_cleaned = []
i = 0
for link in jobs_list:
    if (i%2 == 0):
        print(link[0])
        print(link[1])
        jobs_list_cleaned.append(link[1])
    i+=1
```

```python
jobs_list_cleaned
```

```python
import requests

data = []
i = 0
for url in jobs_list_cleaned:
    print(i)
    article_dict = {}
#     url = jobs_list_cleaned[3]
    # print(url)
    res = requests.get(url) #連線
    soup = BeautifulSoup(res.text, 'lxml')
    content_raw = soup.select('body > div.all > div.skin_middle > div.content > div.left > div.comp_detail')[0]
    content = content_raw.select('div.rr_box')[0].text
    tools = soup.select('body > div.all > div.skin_middle > div.content > div.left > div.comp_detail')[2].text
    title = soup.select('body > div.all > div.skin_middle > div.jobname > div.jobname_title')[0].text.split('\r')[0]
    article_dict['title'] = title
    article_dict['content'] = content
    article_dict['tools'] = tools
    data.append(article_dict)
    i += 1
```

```python
data[3]
```

```python
json_data = []
import json
for each in data:
    jdata = json.dumps(each,ensure_ascii=False)
    json_data.append(jdata)
import csv
with open('./data/yes123_20170514.csv','w') as f:
    wr = csv.writer(f, quoting=csv.QUOTE_ALL)
    for jdata in json_data:
        wr.writerow(jdata)
```
